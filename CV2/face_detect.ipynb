```python
import cv2

# 画像を読み込む
image = cv2.imread('face1.jpg')

# グレースケールに変換
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 顔検出器を作成
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 顔を検出
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# 検出された顔を四角形で囲む
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

# 結果を表示
cv2.imshow('Faces', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# 画像を読み込む
image = cv2.imread('face2.jpg')

# 顔検出用のカスケード分類器を読み込む
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 画像をグレースケールに変換
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 顔を検出
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# 検出された顔に赤色の罫線を描画
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)

# 結果を表示
cv2.imshow('Faces', image)

# 結果を別名で保存
cv2.imwrite('result2.jpg', image)

# キー入力待ち
cv2.waitKey(0)

# ウィンドウを閉じる
cv2.destroyAllWindows()

```

```python
import cv2

# 画像を読み込む
image = cv2.imread('face3.jpg')

# 顔検出用のカスケード分類器を読み込む
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 画像をグレースケールに変換
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 顔を検出
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# 検出された顔に赤色の罫線を描画
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)
    print(f'Face detected at x={x}, y={y}, w={w}, h={h}')

# 結果を表示
cv2.imshow('Faces', image)

# 結果を別名で保存
cv2.imwrite('result3.jpg', image)

# キー入力待ち
cv2.waitKey(0)

# ウィンドウを閉じる
cv2.destroyAllWindows()
```
