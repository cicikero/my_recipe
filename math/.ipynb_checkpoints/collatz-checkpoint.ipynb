```python
import time

# 処理開始時間を記録する
start_time = time.time()

def collatz_conjecture(n):
    count = 0
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        count += 1
    return count

def verify_collatz_conjecture(limit):
    for i in range(1, limit + 1):
        steps = collatz_conjecture(i)
        print("初期値: {}, ステップ数: {}".format(i, steps))

# 上限値を設定
limit = 100

# コラッツ予想を検証
verify_collatz_conjecture(limit)

# 処理終了時間を記録する
end_time = time.time()

# 処理時間を計算する
elapsed_time = end_time - start_time

# 処理時間と出力を表示する
print("処理時間：{:.3f}秒".format(elapsed_time))
```
